```r
library(httr)
library(ggplot2)


a <- httr::GET('http://192.168.1.111:8000/inspectorbudget/entries/?year=2019&limit=10000', accept_json())
request <- content(a, "parsed")
data <- do.call(rbind,lapply(request$results, function(x) as.data.frame(x, stringsAsFactors=FALSE)))
                             year <- sapply(data$date, function(x) substr(x, 1, 4))
month <- sapply(data$date, function(x) substr(x, 6, 7))
data[, c("year", "month")] <- cbind(year, month)
data$amount <- as.numeric(data$amount)
```

```r
cat <- httr::GET('http://192.168.1.111:8000/inspectorbudget/categories/', accept_json())
cat <- content(cat, 'parsed')
cat <- do.call(rbind,lapply(cat$results, function(x) as.data.frame(x, stringsAsFactors=FALSE)))
```

```r
# httr::DELETE('http://192.168.1.111:8000/inspectorbudget/entries/1937')
```

```r
data$x <- paste0(data$year, '-', data$month)
data <- data[data$amount < 0,]
agg <- aggregate(amount ~ x + category, data, sum)
# agg <- agg[agg$sign < 0,]
```

```r
agg <- merge(agg, cat, by.x = 'category', by.y = 'id')
```
